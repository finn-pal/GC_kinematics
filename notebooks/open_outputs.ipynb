```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from astropy.io import ascii
```

```python
it = 0

raw_dir = "/Users/z5114326/Documents/GitHub/GC_kinematics/data/result/m12i/raw/it_%d/" % it
main_out = "allcat_s-%d_p2-7_p3-1.txt" % it
disrupt = "allcat_s-%d_p2-7_p3-1_k-1.5_logm_snap596.txt" % it
gc_id = "allcat_s-%d_p2-7_p3-1_gcid.txt" % it
```

```python
with open(raw_dir + main_out) as f:
    content = f.readlines()

dat = ascii.read(content)

dat["col1"].name = "halo(z=0)"
dat["col2"].name = "logMh(z=0)"
dat["col3"].name = "logM*(z=0)"
dat["col4"].name = "logMh(zfrom)"
dat["col5"].name = "logM*(zform)"
dat["col6"].name = "LogM(tform)"
dat["col7"].name = "zform"
dat["col8"].name = "feh"
dat["col9"].name = "isMPB"
dat["col10"].name = "halo(zform)"
dat["col11"].name = "snapnum(zform)"

main_df = dat.to_pandas()
```

```python
dis = pd.read_csv(raw_dir + disrupt, header=None)
dis.columns = ["log(M_gc)"]
```

```python
gcid = pd.read_csv(raw_dir + gc_id, sep=" ").drop(["ID"], axis=1)
gcid.columns = ["GC_ID", "quality"]
```
